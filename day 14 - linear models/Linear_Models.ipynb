```python
X = data[["athletes", "prev_medals"]]
```

```python
y = data[["medals"]]
```

```python
x["intercept"] = 1
```

```python
x
```

```python
x_T = x.T
```

```python
x_T
```

```python
import numpy as np
```

```python
o = np.linalg.inv(x_T @ x) @ x_T @ y
```

```python
o.index = x.columns
```

```python
o
```

```python
predictions = x @ o
```
