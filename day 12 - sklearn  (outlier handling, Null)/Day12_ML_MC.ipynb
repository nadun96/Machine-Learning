```python
import numpy as np
def zscore(data, col):
  outlier_list = []
  data = data[col]
  mean = np.mean(data)
  std = np.std(data)
  for i in range(data.shape[0]):
    d = data.loc[i]
    z = (d-mean)/std
    if z > 3:
      outlier_list.append(i)
      print("index:", i,"z-score: " ,z)
  print(outlier_list)
```

```python
zscore(data, "Age")
```

```python
def iqr_outlier(data,col):
  q1 = data[col].quantile(0.25)
  q3 = data[col].quantile(0.75)
  iqr = q3 - q1
  lower = q1 - (1.5*iqr)
  upper = q3 + (1.5*iqr)
  data = data[col]
  print(data[(data<lower)|(data>upper)])
```

```python
iqr_outlier(data, "Fare")
```
