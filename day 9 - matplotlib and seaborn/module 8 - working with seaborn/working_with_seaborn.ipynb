```python
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
iris.species.value_counts()
```

```python
sns.set_theme(style="dark")
```

```python
sns.scatterplot(x="sepal_length", y="petal_length", hue="species", size="petal_width",data=iris, palette="rainbow_r")
```

```python
titanic = sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
sns.barplot(x="sex", y="survived",hue="class", data=titanic, palette="RdBu")
```

```python
sns.countplot(x="embarked", data=titanic)
```

```python
iris.head()
```

```python
sns.boxplot(x="species", y="sepal_width", data=iris)
```

```python
sns.pairplot(iris, hue="species", palette="Set1")
```

```python
iris.corr()
```

```python
titanic.corr()
```

```python
sns.heatmap(titanic.corr(), cbar= True, linewidth=2,annot=True, fmt=".1f", cmap="Blues" )
```
