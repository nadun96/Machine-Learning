```python
import pandas as pd
```

```python
heart = pd.read_csv("/content/heart.csv")
```

```python
heart.head()
```

```python
plt.scatter(x=heart.age[heart.target==0], y=heart.chol[heart.target==0], label="Heart disease absent")
plt.scatter(x=heart.age[heart.target==1], y=heart.chol[heart.target==1], label="Heart disease present")
plt.title("age vs cholesterol diagnosis")
plt.xlabel("age")
plt.ylabel("cholesterol")
plt.legend()
plt.show()
```

```python
diabetes = pd.read_csv("/content/diabetes.csv")
```

```python
diabetes.head()
```

```python
nos = diabetes["Outcome"].value_counts()
```

```python
nos
```

```python
print(nos.index)
print(nos.values)
```

```python
plt.bar(nos.index, nos.values)
plt.title("Distribution of OUtcome")
plt.xlabel("Diagnosis")
plt.ylabel("Count")
plt.xticks([0,1],[0,1])


plt.show()
```

```python
diabetes["Age"]
```

```python
plt.hist(diabetes["Age"], bins=4)
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.boxplot([diabetes[diabetes.Outcome == 0].Age, diabetes[diabetes.Outcome == 1].Age], labels=["No diabetes","Diabetes"])
plt.title("Age against diagnosis")
plt.xlabel("Diagnosis")
plt.ylabel("Age")
plt.show()
```
