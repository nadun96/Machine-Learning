```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
```

```python
one = data[data["Class"]==1]
```

```python
zero = data[data["Class"]==0].sample(n=473)
```

```python
df = pd.concat((one,zero), axis=0, ignore_index=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.Class.value_counts()
```

```python
sns.countplot(data=df, x="Class")
```

```python
x = df.drop(["Class"], axis=1)
y = df.Class
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier()
xgb.fit(x_train, y_train)
```

```python
ans = xgb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy_score(y_test, ans)*100
```

```python
sns.heatmap(confusion_matrix(y_test, ans), fmt=".1f", annot=True)
```

```python
from sklearn.model_selection import cross_val_score

cvs = cross_val_score(xgb, x,y, cv=5)
```

```python
print((sum(cvs)/5)*100)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
cvs = cross_val_score(lr, x,y, cv=5)
print((sum(cvs)/5)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
cvs = cross_val_score(rf, x,y, cv=5)
print((sum(cvs)/5)*100)
```
